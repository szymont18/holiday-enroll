```python
from genetics.genetic_solver_2 import Genetic as Genetic_2
from genetics.genetic_solver import GeneticSolver as Genetic
from brute_force.brute_force import BruteForce
from let_it_bee.bee_solver import BeeSolver
from České_žíhání.annealing import Annealing
import matplotlib.pyplot as plt
```

```python
def plot_test(path, no_of_gens):    
    _, values_genetic = Genetic(path, generation_no = no_of_gens).solve()
    _, values_genetic_2 = Genetic_2(path, generation_no = no_of_gens).solve()
#     _, values_brute = BruteForce(path).solve()
#     values_brute = [values_brute for _ in range(len(values_genetic))]
    _, values_bee = BeeSolver(path, generation_no = no_of_gens).solve()
    _, values_ann = Annealing(path, iterations = no_of_gens).solve()
    xes = [i*10 for i in range(0, no_of_gens//10)]
    plt.plot(xes, values_genetic, ".", label = "wyniki algorytmu genetycznego 1", color= "blue")
    plt.plot(xes, values_genetic_2, ".", label = "wyniki algorytmu genetycznego 2", color= "green")
#     plt.plot(xes, values_brute, ".", label = "wyniki algorytmu brute force", color= "black")
    plt.plot(xes, values_bee, ".", label = "wyniki algorytmu pszczelego", color= "orange")
    plt.plot(xes, values_ann, ".", label = "wyniki algorytmu wyżarzania", color= "red")
    plt.legend()
    plt.xlabel("Numer generacji")
    plt.ylabel("Wartość funkcji kosztu")
    plt.title(f"Zależność wartości funkcji kosztu od numeru generacji dla testu {path}")
    ax = plt.gca()
    ax.set_ylim([-300, 2000])
    plt.show()




```

```python
plot_test("tests/test1.json", 10000)
```
